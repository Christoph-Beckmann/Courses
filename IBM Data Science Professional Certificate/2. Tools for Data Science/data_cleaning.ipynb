```python
import csv
import pandas as pd
import numpy as np

df = pd.read_csv("data_heart_cleveland.csv")
```

```python
# Get overview about variables
df.dtypes
```

```python
df.head(5)
```

```python
# remove uneccessary columns

to_drop = ["cp",
           "fbs",
           "restecg",
           "thalach",
           "exang",
           "oldpeak",
           "slope",
           "ca",
           "thal",
           "condition"
           ]

df.drop(to_drop, axis=1, inplace=True)
```

```python
df.head(5)
```

```python
# renaming columns

renaming = {
    "age" : "Age",
    "sex" : "Sex",
    "trestbps" : "Bps",
    "chol" : "Cholesterol"
}

df.rename(columns=renaming, inplace=True) # inplace mean their is no dataframe which is returned
df.head(5)
```

```python
# replacing rows if necessary

replace_values = {0 : "F", 1 : "M"}
df.replace({"Sex": replace_values},inplace=True)
```
